# Data Preprocessing process

#### Let's keep all input feature 

## Separate the data and label 

```python
x = df.drop('quality', axis = 1)
x.head()
```

- if the quality of wine is 6 or less then 6 then it is bad quality else it is good.

#### Label Binarization

```python
y = df['quality'].apply(lambda y_value: 1 if y_value >= 7 else 0)
```

```python
print(y)
```

```python
y.value_counts()
```

# Train & test Split

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 8)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

# Model training 

- Random forest Classifier

```python
model = RandomForestClassifier()
```

```python
model.fit(x_train, y_train)
```

# model Evaluation

```python
# for training data 
pred_xtrain = model.predict(x_train)
print(pred_xtrain)
```

```python
# for test data 
pred_xtest = model.predict(x_test)
print(pred_xtest)
```

```python
accuracy_train = accuracy_score(pred_xtrain,y_train)
accuracy_test = accuracy_score(pred_xtest, y_test)
print("Accuracy of training data:", accuracy_train)
print("Accuracy of testing data:", accuracy_test)
```

```python
cm = confusion_matrix(pred_xtest, y_test)
print(cm)
```

```python
plt.figure(figsize = (6,6))
disp = ConfusionMatrixDisplay(cm, display_labels = ["Bad", "Good"] )
disp.plot(cmap=plt.cm.Blues)
plt.show()
```

```python
report= classification_report(pred_xtest, y_test)
```

```python
print(report)
```
